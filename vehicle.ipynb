```python
import json
```

```python


class Vehicle:
    def __init__(self, plate_number, owner, engine_cc, base_tax):

        if not isinstance(plate_number, str):
            raise TypeError("plate_number must be a string")

        if not isinstance(owner, str):
            raise TypeError("owner must be a string")

        if not isinstance(engine_cc, (int, float)) or engine_cc <= 0:
            raise ValueError("engine_cc must be a positive number")

        if not isinstance(base_tax, (int, float)) or base_tax < 0:
            raise ValueError("base_tax must be a positive number")

        self.plate_number = plate_number
        self.owner = owner
        self.engine_cc = engine_cc
        self.base_tax = base_tax


    def display_info(self):
        return {
            "plate_number": self.plate_number,
            "owner": self.owner,
            "engine_cc": self.engine_cc,
            "base_tax": self.base_tax
        }

    def print_info(self):
        info = self.display_info()
        for key, value in info.items():
            print(f"{key}: {value}")

class Car(Vehicle):
    def __init__(self, plate_number, owner, engine_cc, base_tax, passenger_capacity):

        if not isinstance(passenger_capacity, int) or passenger_capacity <= 0:
            raise ValueError("passenger_capacity must be a positive integer")

        super().__init__(plate_number, owner, engine_cc, base_tax)
        self.passenger_capacity = passenger_capacity

    def calculate_tax(self):
        return self.base_tax + (self.engine_cc * 0.05)

    def display_info(self):
        info = super().display_info()
        info["passenger_capacity"] = self.passenger_capacity
        info["calculated_tax"] = int(self.calculate_tax())
        return info


class Truck(Vehicle):
    def __init__(self, plate_number, owner, engine_cc, base_tax, load_capacity):

        if not isinstance(load_capacity, (int, float)) or load_capacity <= 0:
            raise ValueError("load_capacity must be a positive numeric value")

        super().__init__(plate_number, owner, engine_cc, base_tax)
        self.load_capacity = load_capacity

    def calculate_tax(self):
        return self.base_tax +(self.load_capacity * 0.1 * 1000000)

    def display_info(self):
        info = super().display_info()
        info["load_capacity"] = self.load_capacity
        info["calculated_tax"] = int(self.calculate_tax())
        return info

class Motorbike(Vehicle):
    def __init__(self, plate_number, owner, engine_cc, base_tax, bike_type):

        if not isinstance(bike_type, str):
            raise TypeError("bike_type must be a string")

        super().__init__(plate_number, owner, engine_cc, base_tax)
        self.bike_type = bike_type

    def calculate_tax(self):
        return self.base_tax + 20000

    def display_info(self):
        info = super().display_info()
        info["type"] = self.bike_type
        info["calculated_tax"] = int(self.calculate_tax())
        return info


car1 = Car("UAZ 183 A","Eng. Sowed",1800,200000,5)
truck1 = Truck("Ubh 203 N","Eng. Stella",2500,500000,3)
bike1 = Motorbike("Ugd 642 E","John",150,60000,"Bajaj CT")

vehicles = [car1, truck1, bike1]

for v in vehicles:
    print("")
    v.print_info()

vehicle_data = [v.display_info() for v in vehicles]

with open("veh.json", "w") as f:
    json.dump(vehicle_data, f, indent=4)
print("")
print("All vehicle records have been saved to 'veh.json'")
print("")
```
